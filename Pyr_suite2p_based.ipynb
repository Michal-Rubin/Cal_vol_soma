# Pyr suite2p-based FR comparison

This notebook runs `pyr_suite2p_based.py` for single-cell and DB-level analysis.
Tune `REAL_FR_SMOOTH_SIGMA_S` to control how much the real FR is smoothed before comparison.

```python
import os
import importlib
import pandas as pd

import pyr_suite2p_based as ps
import pyr_event_cal2_pipeline as pe

importlib.reload(ps)

# Main knobs
REAL_FR_SMOOTH_SIGMA_S = 0.10  # seconds
RESID_Z_K = 3.0
MIN_SEG_DUR_S = 0.30
DB_PATH = r'Z:\Adam-Lab-Shared\Data\Michal_Rubin\Dendrites\PyrLowFR.csv'
OUT_DIR = r"Z:\Adam-Lab-Shared\Data\Michal_Rubin\data summery\2026\Pyr\suite2p_vs_realfr"
os.makedirs(OUT_DIR, exist_ok=True)

print("DB_PATH:", pe.DB_PATH)
print("OUT_DIR:", OUT_DIR)
```

```python
# Single-cell example (first row in DB)
db = pd.read_csv(pe.DB_PATH)
row = db.iloc[0]

CELL_FOLDER = str(row["Link"])
CAL_SR = float(row["CALsr"]) if ("CALsr" in row and pd.notna(row["CALsr"])) else None

res = ps.analyze_suite2p_vs_real_fr(
    cell_folder=CELL_FOLDER,
    cal_sr_hz=CAL_SR,
    real_fr_smooth_sigma_s=REAL_FR_SMOOTH_SIGMA_S,
    resid_z_k=RESID_Z_K,
    min_seg_dur_s=MIN_SEG_DUR_S,
    out_dir=OUT_DIR,
    save_outputs=True,
)

print("pearson_r:", res["pearson_r"])
print("FR plot:", res["figure_fr_html"])
print("Event-area plot:", res["figure_events_html"])
print("Framewise CSV:", res["compare_csv"])
```
