```python
# Event-shape summary by event type and spike count (z-score + robust dF/F + local calcium dF/F)
import io
import os
import pickle
import base64
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

import pyr_event_cal2_pipeline as pe

OUT_DIR = Path(r'Z:\Adam-Lab-Shared\Data\Michal_Rubin\data summery\2026\Pyr\event_shape')
OUT_DIR.mkdir(parents=True, exist_ok=True)

VOL_SR = 500.0
PRE_MS = 50.0
POST_MS = 500.0
SPIKE_CAP = 6
DT_MS = 1000.0 / VOL_SR  # 2 ms
T_REL_MS = np.arange(-PRE_MS, POST_MS + 1e-9, DT_MS)
T_REL_S = T_REL_MS / 1000.0
PRE_MASK = T_REL_MS < 0

V_COLOR_SIMPLE = '#f08080'   # light red
V_COLOR_COMPLEX = '#8b0000'  # dark red
C_COLOR_MEAN = '#4b0082'     # dark purple
C_COLOR_INDIV = '#c9a0ff'    # light purple


def _safe_zscore(x):
    x = np.asarray(x, dtype=float)
    mu = np.nanmean(x)
    sd = np.nanstd(x)
    if not np.isfinite(sd) or sd <= 0:
        return np.zeros_like(x, dtype=float)
    return (x - mu) / sd

NON_ROBUST_CAL_LOW_PERCENTILE = 15.0


def _trace_z_robust_p8mad(trace_vol, trace_cal, class_events, cal_sr):
    trace_vol = np.asarray(trace_vol, dtype=float).ravel()
    trace_cal = np.asarray(trace_cal, dtype=float).ravel()

    # Voltage robust z: center/scale from quiet voltage windows (same as pipeline)
    qv, _ = pe._build_quiet_masks(
        trace_vol_len=trace_vol.size,
        trace_cal_len=trace_cal.size,
        class_events=class_events,
        vol_sr=VOL_SR,
        cal_sr=cal_sr,
    )
    if np.any(qv):
        v_center, v_scale = pe._robust_center_scale(trace_vol[qv], center=None)
    else:
        v_center, v_scale = pe._robust_center_scale(trace_vol, center=None)

    if np.isfinite(v_scale) and v_scale > 0 and np.isfinite(v_center):
        vol_z = (trace_vol - float(v_center)) / float(v_scale)
    else:
        vol_z = np.zeros_like(trace_vol, dtype=float)

    # Calcium robust z: 8th percentile center + robust scale (MAD)
    c_center, c_scale = pe._calcium_p8_center_scale(trace_cal, p=8.0)
    if np.isfinite(c_scale) and c_scale > 0 and np.isfinite(c_center):
        cal_z = (trace_cal - float(c_center)) / float(c_scale)
    else:
        cal_z = np.zeros_like(trace_cal, dtype=float)

    return np.asarray(vol_z, dtype=float), np.asarray(cal_z, dtype=float)


def _trace_z_nonrobust_p15(trace_vol, trace_cal, class_events, cal_sr, low_percentile=NON_ROBUST_CAL_LOW_PERCENTILE):
    trace_vol = np.asarray(trace_vol, dtype=float).ravel()
    trace_cal = np.asarray(trace_cal, dtype=float).ravel()

    # Keep voltage robust-z definition as before (quiet-based)
    qv, _ = pe._build_quiet_masks(
        trace_vol_len=trace_vol.size,
        trace_cal_len=trace_cal.size,
        class_events=class_events,
        vol_sr=VOL_SR,
        cal_sr=cal_sr,
    )
    if np.any(qv):
        v_center, v_scale = pe._robust_center_scale(trace_vol[qv], center=None)
    else:
        v_center, v_scale = pe._robust_center_scale(trace_vol, center=None)

    if np.isfinite(v_scale) and v_scale > 0 and np.isfinite(v_center):
        vol_z = (trace_vol - float(v_center)) / float(v_scale)
    else:
        vol_z = np.zeros_like(trace_vol, dtype=float)

    # Calcium non-robust z: center/std from low-percentile population (p15 requested)
    c_center, c_scale = pe._zscore_low8(trace_cal, low_percentile=float(low_percentile))
    if np.isfinite(c_scale) and c_scale > 0 and np.isfinite(c_center):
        cal_z = (trace_cal - float(c_center)) / float(c_scale)
    else:
        cal_z = np.zeros_like(trace_cal, dtype=float)

    return np.asarray(vol_z, dtype=float), np.asarray(cal_z, dtype=float)

def _cap_spike_count(n, cap=SPIKE_CAP):
    try:
        v = int(n)
    except Exception:
        return int(cap)
    return int(min(v, int(cap)))

def _save_fig_pdf_html(fig, fig_title, file_stem, out_dir):
    pdf_path = out_dir / f"{file_stem}.pdf"
    html_path = out_dir / f"{file_stem}.html"

    fig.savefig(pdf_path, format='pdf', bbox_inches='tight')

    buf = io.BytesIO()
    fig.savefig(buf, format='png', dpi=170, bbox_inches='tight')
    buf.seek(0)
    img_b64 = base64.b64encode(buf.read()).decode('ascii')
    buf.close()

    html = f"""<!doctype html>
<html>
<head><meta charset='utf-8'><title>{fig_title}</title></head>
<body>
  <h2>{fig_title}</h2>
  <img src='data:image/png;base64,{img_b64}' style='width:100%; height:auto;' />
</body>
</html>
"""
    html_path.write_text(html, encoding='utf-8')
    print('Saved:', pdf_path)
    print('Saved:', html_path)

def _build_event_rows_for_scaling(class_events, trace_cal, cal_sr, vol_sr):
    rows = []
    nc = int(len(trace_cal))
    if nc <= 0:
        return rows

    for i, ev in enumerate(class_events):
        sp = pe._as_sorted_unique_int(ev.get('spikes', []))
        if sp.size == 0:
            continue

        v_start = int(ev.get('start_frame', int(sp[0])))
        c_start = pe._vol_idx_to_cal_idx(v_start, vol_sr=vol_sr, cal_sr=cal_sr, cal_len=nc)
        c_last_sp = pe._vol_idx_to_cal_idx(int(sp[-1]), vol_sr=vol_sr, cal_sr=cal_sr, cal_len=nc)

        c_next_start = None
        if i < len(class_events) - 1:
            next_ev = class_events[i + 1]
            next_v_start = int(next_ev.get('start_frame', int(sp[-1])))
            c_next_start = pe._vol_idx_to_cal_idx(next_v_start, vol_sr=vol_sr, cal_sr=cal_sr, cal_len=nc)

        peak_n = max(1, int(round(0.300 * float(cal_sr))))
        search_start = int(max(0, min(nc - 1, c_last_sp)))
        search_end = int(min(nc - 1, search_start + peak_n))
        if c_next_start is not None:
            search_end = int(min(search_end, max(search_start, c_next_start - 1)))

        if search_end < search_start:
            peak_idx = search_start
        else:
            seg = np.asarray(trace_cal[search_start:search_end + 1], dtype=float)
            if seg.size == 0 or not np.any(np.isfinite(seg)):
                peak_idx = search_start
            else:
                peak_idx = int(search_start + np.nanargmax(seg))

        rows.append({
            'event_type': str(ev.get('event_type', 'simple')),
            'spikes': sp,
            'start_idx': int(c_start),
            'peak_idx': int(peak_idx),
        })

    return rows

def _norm_path_key(path_like):
    return os.path.normpath(str(path_like)).replace('/', '\\').lower()

def _load_chosen_event_idx_map(summary_csv_path):
    idx_map = {}
    if not summary_csv_path:
        return idx_map
    if not os.path.isfile(summary_csv_path):
        return idx_map
    try:
        df = pd.read_csv(summary_csv_path)
    except Exception:
        return idx_map

    needed = {'cell_folder', 'pkl_name', 'event_idx'}
    if not needed.issubset(set(df.columns)):
        return idx_map

    df = df.dropna(subset=['cell_folder', 'pkl_name', 'event_idx']).copy()
    if len(df) == 0:
        return idx_map

    for (cell_folder, pkl_name), sub in df.groupby(['cell_folder', 'pkl_name']):
        try:
            ev_idx = set(sub['event_idx'].astype(int).tolist())
        except Exception:
            continue
        idx_map[(_norm_path_key(cell_folder), str(pkl_name).lower())] = ev_idx
    return idx_map

def _build_pipeline_event_rows_for_shapes(d, trace_vol, trace_cal, cal_sr, vol_sr=VOL_SR):
    class_events, _, _ = pe._events_from_saved_labels(
        d,
        trace_len=trace_vol.size,
        trace_vol=trace_vol,
        vol_sr=vol_sr,
        cs_z_start_thr=pe.CS_Z_START_THR,
        cs_z_end_thr=pe.CS_Z_END_THR,
        non_cs_pad_frames=5,
        simple_isi_ms=pe.SIMPLE_ISI_MS,
    )
    if len(class_events) == 0:
        return [], []

    class_events = sorted(class_events, key=lambda ev: int(ev.get('start_frame', 0)))
    cal_t = np.arange(trace_cal.size, dtype=float) / float(cal_sr)
    mu_low8, sd_low8 = pe._zscore_low8(trace_cal, low_percentile=pe.CAL_Z_NON_ROBUST_LOW_PERCENTILE)

    usable_events = []
    for ev in class_events:
        ev_spk = pe._as_sorted_unique_int(ev.get('spikes', []))
        ev_spk = ev_spk[(ev_spk >= 0) & (ev_spk < trace_vol.size)]
        if ev_spk.size == 0:
            continue
        ev_start = int(max(0, min(trace_vol.size - 1, int(ev.get('start_frame', int(ev_spk[0]))))))
        usable_events.append((ev_start, int(ev_spk[0]), ev, ev_spk))

    usable_events = sorted(usable_events, key=lambda x: int(x[1]))
    event_rows = []

    for i, (_ev_start, _first_sp, ev, ev_spk) in enumerate(usable_events):
        v_start = int(max(0, min(trace_vol.size - 1, int(ev.get('start_frame', int(ev_spk[0]))))))
        v_end = int(max(v_start, min(trace_vol.size - 1, int(ev.get('end_frame', int(ev_spk[-1]))))))

        c_start = pe._vol_idx_to_cal_idx(v_start, vol_sr=vol_sr, cal_sr=cal_sr, cal_len=trace_cal.size)
        c_first_sp = pe._vol_idx_to_cal_idx(int(ev_spk[0]), vol_sr=vol_sr, cal_sr=cal_sr, cal_len=trace_cal.size)
        c_last_sp = pe._vol_idx_to_cal_idx(int(ev_spk[-1]), vol_sr=vol_sr, cal_sr=cal_sr, cal_len=trace_cal.size)

        c_next_start = None
        c_next_start_for_feat = None
        if i < len(usable_events) - 1:
            _next_start, _next_first_sp, next_ev, next_spk = usable_events[i + 1]
            if next_spk is not None and np.asarray(next_spk).size > 0:
                next_v_start = int(max(0, min(trace_vol.size - 1, int(next_spk[0]))))
            else:
                next_v_start = int(max(0, min(trace_vol.size - 1, int(next_ev.get('start_frame', _next_start)))))

            curr_last_sp = int(ev_spk[-1])
            if next_v_start > curr_last_sp:
                if v_end >= next_v_start:
                    v_end = int(max(v_start, next_v_start - 1))
                c_next_start = pe._vol_idx_to_cal_idx(next_v_start, vol_sr=vol_sr, cal_sr=cal_sr, cal_len=trace_cal.size)
                c_next_start_for_feat = c_next_start

        c_end = pe._vol_idx_to_cal_idx(v_end, vol_sr=vol_sr, cal_sr=cal_sr, cal_len=trace_cal.size)
        c_end = int(max(c_end, c_last_sp))

        if c_next_start is not None:
            c_end_cap = int(max(c_start, min(trace_cal.size - 1, c_next_start - 1)))
            if c_end > c_end_cap and c_end_cap > c_start:
                c_end = c_end_cap
            elif c_end_cap <= c_start:
                c_end = int(c_start)
                c_next_start_for_feat = int(min(trace_cal.size - 1, c_start + 1))

        is_complex_event = (str(ev.get('event_type', '')) == 'complex') or (str(ev.get('event_kind', '')) == 'complex')
        is_burst_event = (str(ev.get('event_kind', '')) == 'simple_burst') or is_complex_event

        peak_window_s = float(pe.COMPLEX_PEAK_SEARCH_S if is_complex_event else pe.SIMPLE_PEAK_SEARCH_S)
        peak_search_start_idx = int(c_first_sp)
        peak_n = max(1, int(round(float(peak_window_s) * float(cal_sr))))
        peak_search_end_idx = int(min(trace_cal.size - 1, peak_search_start_idx + peak_n))

        if c_next_start_for_feat is not None:
            peak_search_end_idx = int(min(peak_search_end_idx, max(peak_search_start_idx, c_next_start_for_feat - 1)))

        c_end_for_feat = int(max(c_end, peak_search_end_idx))
        if c_next_start_for_feat is not None:
            c_end_for_feat = int(min(c_end_for_feat, max(c_start, c_next_start_for_feat - 1)))

        feat = pe._calc_event_features(
            trace_cal,
            cal_t,
            start_idx=c_start,
            end_idx=c_end_for_feat,
            next_start_idx=c_next_start_for_feat,
            mu_low8=mu_low8,
            sd_low8=sd_low8,
            max_peak_search_s=peak_window_s,
            pre_baseline_s=pe.PRE_BASELINE_S,
            end_frac=pe.EVENT_END_FRAC,
            use_baseline_frac_end=is_burst_event,
            peak_search_start_idx=peak_search_start_idx,
            peak_search_end_idx=peak_search_end_idx,
            first_spike_idx=c_first_sp,
        )
        if feat is None:
            continue

        event_rows.append({
            'event_idx': len(event_rows),
            'spikes': ev_spk,
            'n_spikes': int(ev_spk.size),
            'event_type': str(ev.get('event_type', 'simple')),
            'event_kind': str(ev.get('event_kind', 'single')),
            'bound_method': str(ev.get('bound_method', '')),
            'start_frame_v': int(v_start),
            'end_frame_v': int(v_end),
            'is_burst_event': bool(is_burst_event),
            'is_complex_event': bool(is_complex_event),
            **feat,
        })

    if len(event_rows) == 0:
        return class_events, []

    event_rows, _ = pe._merge_events_same_start(event_rows)
    event_rows = pe._evaluate_include_rule(
        event_rows,
        trace_cal=trace_cal,
        ratio_thr=pe.TAIL_RATIO_THR,
        cal_sr=cal_sr,
        ratio_max_gap_simple_s=pe.GAP_THR_PREV_SIMPLE_S,
        ratio_max_gap_complex_s=pe.GAP_THR_PREV_COMPLEX_S,
        vol_sr=vol_sr,
    )
    event_rows, _ = pe._resolve_duplicate_event_peaks(event_rows, tol_frames=pe.PEAK_DUP_TOL_FRAMES)
    return class_events, event_rows

def _collect_event_windows(mode='zscore', include_only=('simple', 'complex')):
    groups = {}

    chosen_idx_map = _load_chosen_event_idx_map(getattr(pe, 'SUMMARY_CSV', None))

    for ridx, row in DB.iterrows():
        cell_folder = str(row['Link'])
        cal_sr = pe._safe_cal_sr(row.get('CALsr', 30.0), default=30.0)

        if not os.path.isdir(cell_folder):
            continue

        pkl_list = pe._find_spike_pkls(cell_folder)
        if len(pkl_list) == 0:
            continue

        main_pkl = None
        for pkl in pkl_list:
            if pe._suffix_from_pkl_name(pkl) == 'main':
                main_pkl = pkl
                break
        if main_pkl is None:
            main_pkl = pkl_list[0]

        try:
            with open(main_pkl, 'rb') as f:
                d = pickle.load(f)
        except Exception:
            continue

        trace_vol = np.asarray(d.get('trace_vol', []), dtype=float).ravel()
        trace_cal = np.asarray(d.get('trace_cal', []), dtype=float).ravel()
        if trace_vol.size == 0 or trace_cal.size == 0:
            continue

        class_events, event_rows = _build_pipeline_event_rows_for_shapes(
            d=d,
            trace_vol=trace_vol,
            trace_cal=trace_cal,
            cal_sr=cal_sr,
            vol_sr=VOL_SR,
        )
        if len(class_events) == 0 or len(event_rows) == 0:
            continue

        map_key = (_norm_path_key(cell_folder), os.path.basename(main_pkl).lower())
        chosen_idx = chosen_idx_map.get(map_key)
        if chosen_idx is not None and len(chosen_idx) > 0:
            chosen_events = [ev for ev in event_rows if int(ev.get('event_idx', -1)) in chosen_idx]
        else:
            chosen_events = [ev for ev in event_rows if bool(ev.get('include', False))]

        if len(chosen_events) == 0:
            continue
        if mode == 'zscore_nonrobust_p15':
            vol_sig, cal_sig = _trace_z_nonrobust_p15(
                trace_vol=trace_vol,
                trace_cal=trace_cal,
                class_events=class_events,
                cal_sr=cal_sr,
                low_percentile=NON_ROBUST_CAL_LOW_PERCENTILE,
            )
            cal_sig_raw = None
        elif mode == 'zscore_robust_p8mad':
            vol_sig, cal_sig = _trace_z_robust_p8mad(
                trace_vol=trace_vol,
                trace_cal=trace_cal,
                class_events=class_events,
                cal_sr=cal_sr,
            )
            cal_sig_raw = None
        elif mode == 'robust_dff':
            event_rows_for_scale = _build_event_rows_for_scaling(
                class_events=class_events,
                trace_cal=trace_cal,
                cal_sr=cal_sr,
                vol_sr=VOL_SR,
            )
            scale_info = pe._compute_robust_scaling_from_cs(
                trace_vol=trace_vol,
                trace_cal=trace_cal,
                class_events=class_events,
                event_rows=event_rows_for_scale,
                vol_sr=VOL_SR,
                cal_sr=cal_sr,
                pre_baseline_s=0.5,
            )
            vol_sig = trace_vol * float(scale_info.get('scale_vol', 1.0))
            cal_sig = trace_cal * float(scale_info.get('scale_cal', 1.0))
            cal_sig_raw = None
        elif mode == 'cal_local_dff':
            vol_sig, _cal_z_unused = _trace_z_nonrobust_p15(
                trace_vol=trace_vol,
                trace_cal=trace_cal,
                class_events=class_events,
                cal_sr=cal_sr,
                low_percentile=NON_ROBUST_CAL_LOW_PERCENTILE,
            )
            cal_sig = None
            cal_sig_raw = np.asarray(trace_cal, dtype=float)
        else:
            raise ValueError(f'Unknown mode: {mode}')

        vol_t = np.arange(trace_vol.size, dtype=float) / float(VOL_SR)
        cal_t = np.arange(trace_cal.size, dtype=float) / float(cal_sr)

        for ev in chosen_events:
            ev_type = str(ev.get('event_type', 'simple'))
            if ev_type not in include_only:
                continue

            sp = pe._as_sorted_unique_int(ev.get('spikes', []))
            if sp.size == 0:
                continue

            first_sp = int(sp[0])
            if first_sp < 0 or first_sp >= trace_vol.size:
                continue

            t0 = float(first_sp) / float(VOL_SR)
            tq = t0 + T_REL_S

            vol_evt = np.interp(tq, vol_t, vol_sig, left=np.nan, right=np.nan)

            if mode == 'cal_local_dff':
                cal_evt_raw = np.interp(tq, cal_t, cal_sig_raw, left=np.nan, right=np.nan)
                pre_vals = cal_evt_raw[PRE_MASK]
                pre_vals = pre_vals[np.isfinite(pre_vals)]
                if pre_vals.size < 3:
                    continue
                f0 = float(np.nanpercentile(pre_vals, 8))
                pre_med_abs = float(np.nanpercentile(np.abs(pre_vals), 50)) if pre_vals.size else 0.0
                trace_abs = np.abs(trace_cal[np.isfinite(trace_cal)])
                trace_p95_abs = float(np.nanpercentile(trace_abs, 95)) if trace_abs.size else 0.0
                denom_floor = max(1e-3, pre_med_abs, 0.05 * trace_p95_abs, 0.05)
                denom = max(abs(f0), denom_floor) if np.isfinite(f0) else np.nan
                if not np.isfinite(denom) or denom <= 0:
                    continue
                cal_evt = (cal_evt_raw - f0) / denom
            else:
                cal_evt = np.interp(tq, cal_t, cal_sig, left=np.nan, right=np.nan)

            if np.sum(np.isfinite(vol_evt)) < int(0.8 * vol_evt.size):
                continue
            if np.sum(np.isfinite(cal_evt)) < int(0.5 * cal_evt.size):
                continue

            n_bin = _cap_spike_count(sp.size, cap=SPIKE_CAP)
            key = (ev_type, n_bin)
            if key not in groups:
                groups[key] = {
                    'vol': [],
                    'cal': [],
                    'cell_ids': set(),
                }
            groups[key]['vol'].append(vol_evt)
            groups[key]['cal'].append(cal_evt)
            groups[key]['cell_ids'].add(ridx)

    return groups

def _plot_groups_2rows(groups, mode_title='zscore', file_stem='event_shape', cal_ylim_cap=None, add_capped_rows=True):
    if len(groups) == 0:
        raise RuntimeError(f'No event windows collected for mode={mode_title}.')

    spike_bins = sorted(set(k[1] for k in groups.keys()))
    if len(spike_bins) == 0:
        raise RuntimeError('No spike bins available for plotting.')

    n_cols = len(spike_bins)
    n_rows = 4 if add_capped_rows else 2
    fig_h = 3.8 * float(n_rows)
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(4.8 * n_cols, fig_h), squeeze=False)

    row_types = ['simple', 'complex']

    # Row-wise shared calcium ranges (uncapped rows): same y-range for all subplots in each row.
    row_cal_ylim = {}
    for ev_type in row_types:
        row_vals = []
        for sb in spike_bins:
            entry = groups.get((ev_type, sb), None)
            if entry is None or len(entry.get('cal', [])) == 0:
                continue
            arr = np.asarray(entry['cal'], dtype=float)
            arr = arr[np.isfinite(arr)]
            if arr.size > 0:
                row_vals.append(arr)
        if len(row_vals) == 0:
            row_cal_ylim[ev_type] = (-0.15, 1.0)
        else:
            vals = np.concatenate(row_vals)
            y_min = float(np.nanmin(vals))
            y_max = float(np.nanmax(vals))
            if not np.isfinite(y_min):
                y_min = -0.15
            if not np.isfinite(y_max):
                y_max = 1.0
            y_min = min(-0.15, y_min)
            if y_max <= y_min:
                y_max = y_min + 1.0
            pad = 0.03 * max(1e-6, (y_max - y_min))
            row_cal_ylim[ev_type] = (y_min - pad, y_max + pad)

    def _draw_row(ax_row, ev_type, use_cap=False):
        for c, sb in enumerate(spike_bins):
            ax = axes[ax_row, c]
            ax2 = ax.twinx()
            key = (ev_type, sb)
            entry = groups.get(key, None)

            if use_cap and (cal_ylim_cap is not None):
                cal_ylim = (-0.15, float(cal_ylim_cap))
            else:
                cal_ylim = row_cal_ylim.get(ev_type, (-0.15, 1.0))

            if entry is None or len(entry.get('cal', [])) == 0:
                ax.set_xlim(-PRE_MS, POST_MS)
                ax.axvline(0.0, color='black', linestyle='--', linewidth=1.0, alpha=0.7)
                ax.grid(alpha=0.2)
                title_bin = f'{sb}+' if int(sb) == int(SPIKE_CAP) else str(sb)
                row_suffix = ' (capped)' if use_cap else ''
                ax.set_title(f'{ev_type.capitalize()}{row_suffix} | spikes={title_bin} | events=0 | cells=0', fontsize=10)
                ax.set_xlabel('Time from first spike (ms)')
                if c == 0:
                    vcol = V_COLOR_SIMPLE if ev_type == 'simple' else V_COLOR_COMPLEX
                    ax.set_ylabel('Voltage', color=vcol)
                    ax2.set_ylabel('Calcium', color=C_COLOR_MEAN)
                ax.tick_params(axis='y', colors=(V_COLOR_SIMPLE if ev_type == 'simple' else V_COLOR_COMPLEX))
                ax2.tick_params(axis='y', colors=C_COLOR_MEAN)
                ax2.set_ylim(cal_ylim)
                continue

            vol_arr = np.asarray(entry['vol'], dtype=float)
            cal_arr = np.asarray(entry['cal'], dtype=float)

            vol_mean = np.nanmean(vol_arr, axis=0)
            cal_mean = np.nanmean(cal_arr, axis=0)

            for y in cal_arr:
                ax2.plot(T_REL_MS, y, color=C_COLOR_INDIV, alpha=0.18, linewidth=0.9, zorder=1)

            vcol = V_COLOR_SIMPLE if ev_type == 'simple' else V_COLOR_COMPLEX
            ax.plot(T_REL_MS, vol_mean, color=vcol, linewidth=2.2, zorder=4, label='Voltage mean')
            ax2.plot(T_REL_MS, cal_mean, color=C_COLOR_MEAN, linewidth=2.8, zorder=3, label='Calcium mean')

            ax.axvline(0.0, color='black', linestyle='--', linewidth=1.0, alpha=0.8)
            ax.set_xlim(-PRE_MS, POST_MS)
            ax.grid(alpha=0.22)
            ax2.set_ylim(cal_ylim)

            n_ev = int(cal_arr.shape[0])
            n_cells = int(len(entry['cell_ids']))
            title_bin = f'{sb}+' if int(sb) == int(SPIKE_CAP) else str(sb)
            row_suffix = ' (capped)' if use_cap else ''
            ax.set_title(f'{ev_type.capitalize()}{row_suffix} | spikes={title_bin} | events={n_ev} | cells={n_cells}', fontsize=10)
            ax.set_xlabel('Time from first spike (ms)')

            if c == 0:
                ax.set_ylabel('Voltage', color=vcol)
                ax2.set_ylabel('Calcium', color=C_COLOR_MEAN)
            ax.tick_params(axis='y', labelcolor=vcol)
            ax2.tick_params(axis='y', labelcolor=C_COLOR_MEAN)

            if ax_row == 0 and c == 0:
                h1, l1 = ax.get_legend_handles_labels()
                h2, l2 = ax2.get_legend_handles_labels()
                ax.legend(h1 + h2, l1 + l2, loc='best', fontsize=8)

    # Base two rows (shared row-wise calcium range)
    _draw_row(ax_row=0, ev_type='simple', use_cap=False)
    _draw_row(ax_row=1, ev_type='complex', use_cap=False)

    # Additional two rows with fixed capped calcium ranges
    if add_capped_rows:
        _draw_row(ax_row=2, ev_type='simple', use_cap=True)
        _draw_row(ax_row=3, ev_type='complex', use_cap=True)

    fig.suptitle(
        f'Pyramidal event shapes by type and spike count ({mode_title})\n'
        f'Window: -{int(PRE_MS)} ms to +{int(POST_MS)} ms around first spike (t=0) | spikes {SPIKE_CAP}+ grouped',
        fontsize=13,
    )
    fig.tight_layout(rect=[0, 0, 1, 0.95])

    _save_fig_pdf_html(
        fig,
        fig_title=f'Pyr event shapes by type/spike count ({mode_title})',
        file_stem=file_stem,
        out_dir=OUT_DIR,
    )
    plt.close(fig)

# 1) non-robust z version (calcium low15 center/std)
groups_z_nonrobust = _collect_event_windows(mode='zscore_nonrobust_p15')
_plot_groups_2rows(
    groups=groups_z_nonrobust,
    mode_title='nonrobust_z_traces_low15',
    file_stem='pyr_event_shape_4rows_spikecountcap6_nonrobust_z_low15',
    cal_ylim_cap=3.0,
    add_capped_rows=True,
)

# 2) robust z version (calcium p8+MAD)
groups_z_robust = _collect_event_windows(mode='zscore_robust_p8mad')
_plot_groups_2rows(
    groups=groups_z_robust,
    mode_title='robust_z_traces_p8mad',
    file_stem='pyr_event_shape_4rows_spikecountcap6_robust_z_p8mad',
    cal_ylim_cap=5.0,
    add_capped_rows=True,
)

# 3) robust dF/F-scaled version (trace-level robust scaling)
groups_robust = _collect_event_windows(mode='robust_dff')
_plot_groups_2rows(
    groups=groups_robust,
    mode_title='robust_dff_scaled_traces',
    file_stem='pyr_event_shape_4rows_spikecountcap6_robust_dff_scaled',
    cal_ylim_cap=4.0,
    add_capped_rows=True,
)

# 4) additional figure: calcium local dF/F (F0=8th percentile in -50..0 ms pre-spike)
groups_local = _collect_event_windows(mode='cal_local_dff')
_plot_groups_2rows(
    groups=groups_local,
    mode_title='voltage_zscore_calcium_local_dff_pre8pct',
    file_stem='pyr_event_shape_4rows_spikecountcap6_voltage_zscore_calcium_local_dff',
    cal_ylim_cap=4.0,
    add_capped_rows=True,
)

print('Done. Saved figures to:', OUT_DIR)
```
